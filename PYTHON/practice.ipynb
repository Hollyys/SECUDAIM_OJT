```python
def level(hp):
    if hp > 500:
        return 'super'
    elif hp > 300:
        return 'high'
    else:
        return 'normal'
```

```python
# df['level'] = np.where(df['hp'] > 500, 'super', 'normal')
df['level'] = df['hp'].apply(level)
df
```

```python
def explanation(row):
    return row.maker + ' ' + row.model  + ': ' + row.level + ' performance car.'
```

```python
df['explanation'] = df.apply(explanation, axis = 1)
df
```
